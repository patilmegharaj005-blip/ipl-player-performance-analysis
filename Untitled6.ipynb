```python
df.describe()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['runs'], bins=20, kde=True)
plt.title("Runs Distribution")
plt.xlabel("Runs")
plt.ylabel("Frequency")
plt.show()
```

```python
top_runs = df.sort_values(by='runs', ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_runs['runs'], y=top_runs['player'])
plt.title("Top 10 Players by Runs")
plt.xlabel("Runs")
plt.ylabel("Players")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='sr', y='runs', data=df)
plt.title("Strike Rate vs Runs")
plt.xlabel("Strike Rate")
plt.ylabel("Runs")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='ave', y='runs', data=df)
plt.title("Average vs Runs")
plt.xlabel("Average")
plt.ylabel("Runs")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='fours', y='sixes', data=df)
plt.title("4s vs 6s")
plt.xlabel("Fours")
plt.ylabel("Sixes")
plt.show()
```

```python
season_runs = df.groupby('season')['runs'].sum()

plt.figure(figsize=(10,5))
season_runs.plot(kind='line')
plt.title("Season-wise Total Runs")
plt.xlabel("Season")
plt.ylabel("Runs")
plt.show()
```

```python
top_100 = df.sort_values(by='centuries', ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_100['centuries'], y=top_100['player'])
plt.title("Top Players with Most Centuries")
plt.xlabel("Centuries")
plt.ylabel("Players")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
